# 1. Bar Chart

```python
cars_df=pd.read_csv('cars.csv')
cars_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,3))
sns.countplot(x='Type',data=cars_df)
plt.title("Countplot of Type of Cars")
plt.grid()
plt.show()
```

# 2. Histogram (Univariate - Numeric Variable):
#Let's consider 'Price' as a numeric variable.

```python
# Univariate Histogram for 'Price'
plt.title(' Histogram of Price')
plt.hist(cars_df['Price'],edgecolor="black")
plt.xlabel('Price')
plt.ylabel('Count')
```

# 3. Pie Chart (Univariate - Categorical Variable):
#Let's consider 'DriveTrain' as a categorical variable.

```python
x=cars_df['Type'].value_counts()
plt.pie(x,labels=x.index,autopct='%0.02f%%')
plt.show()
```
